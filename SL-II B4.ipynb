```python
import numpy as np
```

```python
x1 = np.array([1, -1, 1, -1])
x2 = np.array([-1, 1, -1, 1])
x3 = np.array([1, 1, -1, -1])
x4 = np.array([-1, -1, 1, 1])

patterns = np.array([x1, x2, x3, x4])

n = patterns.shape[1]
```

```python
W = np.zeros((n, n))

for p in patterns:
    p = p.reshape(n, 1)
    W = W + np.dot(p, p.T)

for i in range(n):
    W[i][i] = 0

print("Weight Matrix:")
print(W)
```

```python
test = np.array([1, -1, 1, 1])

print("\nTest Input:")
print(test)
```

```python
for _ in range(5):
    net = np.dot(W, test)
    
    for i in range(len(net)):
        if net[i] > 0:
            test[i] = 1
        else:
            test[i] = -1

print("\nRecovered Pattern:")
print(test)
```
