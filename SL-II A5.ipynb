```python
import numpy as np
```

```python
def sign(x):
    return np.where(x >= 0, 1, -1)
```

```python
X1 = np.array([1, -1, 1])
Y1 = np.array([1, 1])

X2 = np.array([-1, 1, -1])
Y2 = np.array([-1, -1])

X1 = X1.reshape(1, -1)
Y1 = Y1.reshape(1, -1)

X2 = X2.reshape(1, -1)
Y2 = Y2.reshape(1, -1)
```

```python
W = np.dot(X1.T, Y1) + np.dot(X2.T, Y2)

print("Weight Matrix W:\n", W)
```

```python
def bam_recall(X_input, W, max_iter=10):
    X = X_input.copy()
    
    for i in range(max_iter):
        print(f"\nIteration {i+1}")
        
        Y = sign(np.dot(X, W))
        print("Y:", Y)
        
        X_new = sign(np.dot(Y, W.T))
        print("X:", X_new)
        
        if np.array_equal(X, X_new):
            print("Converged!")
            break
        
        X = X_new
    
    return X, Y
```

```python
print("\n--- Testing BAM Recall ---")

X_test = X1
final_X, final_Y = bam_recall(X_test, W)

print("\nFinal Output:")
print("X:", final_X)
print("Y:", final_Y)
```
