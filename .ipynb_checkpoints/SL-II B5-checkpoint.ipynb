```python
import tensorflow as tf
import keras_cv
import numpy as tf_numpy
import matplotlib.pyplot as plt
from PIL import Image
import requests
from io import BytesIO

model = keras_cv.models.RetinaNet.from_preset(
    "retinanet_resnet50_pascalvoc", 
    load_weights=True
)

url = "https://raw.githubusercontent.com/pytorch/vision/main/gallery/assets/dog2.jpg"
response = requests.get(url)
img = Image.open(BytesIO(response.content)).convert("RGB")
img_resized = img.resize((640, 640))
image_array = tf.keras.utils.img_to_array(img_resized)
image_batch = tf.expand_dims(image_array, axis=0)

predictions = model.predict(image_batch)

plt.figure(figsize=(10, 7))
keras_cv.visualization.plot_bounding_box_gallery(
    image_batch,
    value_range=(0, 255),
    rows=1,
    cols=1,
    y_pred=predictions,
    scale=1.0,
    font_scale=0.7,
    bounding_box_format="xywh",
    class_mapping={i: str(i) for i in range(20)} 
)
plt.show()
```
