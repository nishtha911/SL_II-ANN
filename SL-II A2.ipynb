```python
import numpy as np
```

```python
class McCullochPittsNeuron:
    def __init__(self, weights, threshold):
        self.weights = np.array(weights)
        self.threshold = threshold

    def activate(self, inputs):
        linear_combination = np.dot(inputs, self.weights)
        return 1 if linear_combination >= self.threshold else 0
```

```python
weights = [1, -1]
threshold = 1
```

```python
neuron = McCullochPittsNeuron(weights, threshold)
```

```python
test_inputs = [
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
]
```

```python
print("McCulloch-Pitts Neuron: ANDNOT Function")
print("-" * 35)
print("Input (x1, x2) | Output (y)")
print("-" * 35)
```

```python
for inputs in test_inputs:
    output = neuron.activate(inputs)
    print(f"     {inputs}      |      {output}")
```
