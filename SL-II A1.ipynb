```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
```

```python
def relu(x):
    return np.maximum(0, x)
```

```python
def leaky_relu(x, alpha=0.1):
    return np.where(x > 0, x, x * alpha)
```

```python
def tanh(x):
    return np.tanh(x)
```

```python
x = np.linspace(-5, 5, 200)
```

```python
plt.figure(figsize=(12, 8))
```

```python
plt.subplot(2, 2, 1)
plt.plot(x, sigmoid(x), 'b', lw=2)
plt.title('Sigmoid')
plt.grid(True)
```

```python
plt.subplot(2, 2, 2)
plt.plot(x, relu(x), 'r', lw=2)
plt.title('ReLU (Rectified Linear Unit)')
plt.grid(True)
```

```python
plt.subplot(2, 2, 3)
plt.plot(x, tanh(x), 'g', lw=2)
plt.title('Hyperbolic Tangent (Tanh)')
plt.grid(True)
```

```python
plt.subplot(2, 2, 4)
plt.plot(x, leaky_relu(x), 'm', lw=2)
plt.title('Leaky ReLU')
plt.grid(True)
```

```python
plt.tight_layout()
plt.show()
```
