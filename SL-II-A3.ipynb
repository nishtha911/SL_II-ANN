```python
import numpy as np
```

```python
def get_ascii_binary(n):
    return [int(x) for x in format(ord(str(n)), '08b')]
```

```python
numbers = range(10)
X = np.array([get_ascii_binary(n) for n in numbers])
y = np.array([n % 2 for n in numbers])
```

```python
class Perceptron:
    def __init__(self, input_size, lr=0.1, epochs=100):
        self.weights = np.zeros(input_size + 1) # +1 for Bias
        self.lr = lr
        self.epochs = epochs

    def activation(self, x):
        return 1 if x >= 0 else 0

    def predict(self, x):
        z = np.dot(x, self.weights[1:]) + self.weights[0]
        return self.activation(z)

    def train(self, X, y):
        for _ in range(self.epochs):
            for inputs, label in zip(X, y):
                prediction = self.predict(inputs)
                error = label - prediction
                # Update weights and bias
                self.weights[1:] += self.lr * error * inputs
                self.weights[0] += self.lr * error
```

```python
model = Perceptron(input_size=8)
model.train(X, y)
```

```python
print("Testing the Perceptron:")
print("------------------------")
test_digits = ['0', '3', '4', '7', '8']
for char in test_digits:
    binary_input = np.array(get_ascii_binary(char))
    result = model.predict(binary_input)
    parity = "Odd" if result == 1 else "Even"
    print(f"Digit: {char} | ASCII: {ord(char)} | Prediction: {parity}")
```
