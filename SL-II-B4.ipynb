```python
import numpy as np
```

```python
print("Hopfield Network Storing 4 Vectors\n")

x1 = np.array([1, -1, 1, -1])
x2 = np.array([1, 1, -1, -1])
x3 = np.array([-1, 1, 1, -1])
x4 = np.array([-1, -1, 1, 1])
```

```python
patterns = np.array([x1, x2, x3, x4])

print("Stored Vectors:")
for i, p in enumerate(patterns):
    print("Vector", i+1, ":", p)
```

```python
n = patterns.shape[1]
weights = np.zeros((n, n))

for p in patterns:
    p = p.reshape(n, 1)
    weights += np.dot(p, p.T)

np.fill_diagonal(weights, 0)

print("\nWeight Matrix:")
print(weights)
```

```python
def hopfield_recall(test, weights, steps=5):
    state = test.copy()
    for _ in range(steps):
        for i in range(len(state)):
            raw = np.dot(weights[i], state)
            state[i] = 1 if raw >= 0 else -1
    return state
```

```python
print("\nEnter Test Vector Values (1 or -1)")
test_vector = []

for i in range(n):
    value = int(input("Enter value "+str(i+1)+": "))
    test_vector.append(value)

test_vector = np.array(test_vector)
```

```python
print("\nInput Vector:")
print(test_vector)
```

```python
output = hopfield_recall(test_vector, weights)

print("\nRecalled Vector:")
print(output)
```

```python
print("\nComparison With Stored Vectors:")
for i, p in enumerate(patterns):
    print("Vector", i+1, ":", p, " Match:", np.array_equal(output, p))
```
