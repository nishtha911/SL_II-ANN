```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
X = [
    [2,2],
    [1,3],
    [2,1],
    [4,5],
    [5,4],
    [6,5]
]

y = [0,0,0,1,1,1]
```

```python
w1 = 0.0
w2 = 0.0
bias = 0.0

learning_rate = 0.1
epochs = 20
```

```python
def activation(z):
    if z >= 0:
        return 1
    else:
        return 0
```

```python
for epoch in range(epochs):

    for i in range(len(X)):

        x1 = X[i][0]
        x2 = X[i][1]

        z = (w1 * x1) + (w2 * x2) + bias

        y_pred = activation(z)

        error = y[i] - y_pred

        w1 = w1 + learning_rate * error * x1
        w2 = w2 + learning_rate * error * x2

        bias = bias + learning_rate * error

print("Final Weights:", w1, w2)
print("Final Bias:", bias)
```

```python
class0_x = []
class0_y = []

class1_x = []
class1_y = []

for i in range(len(X)):
    
    if y[i] == 0:
        class0_x.append(X[i][0])
        class0_y.append(X[i][1])
    else:
        class1_x.append(X[i][0])
        class1_y.append(X[i][1])

plt.scatter(class0_x, class0_y, color='red', label='Class 0')
plt.scatter(class1_x, class1_y, color='blue', label='Class 1')
```

```python
x_values = np.linspace(0,7,100)

y_values = []

for x in x_values:
    y_val = -(w1*x + bias)/w2
    y_values.append(y_val)

plt.plot(x_values, y_values, 'k--', label="Decision Boundary")

plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Perceptron Learning Law with Decision Boundary")
plt.legend()
plt.show()
```
